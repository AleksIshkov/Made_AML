```python
def check_mask(tournament):
    mask_len = -1
    for team in tournament:
        if 'mask' not in team:
            return False
        
        mask = team['mask']
        if mask is None:
            return False
        
        team_mask_len = len(mask)
        if mask_len == -1:
            mask_len = team_mask_len
        else:
            if mask_len != team_mask_len:
                return False
    return True
```

```python
train_json = {key: value for key, value in results_json.items() if key in train_ids and check_mask(value)}
test_json = {key: value for key, value in results_json.items() if key in test_ids and check_mask(value)}

print('Train tour = {}\nTest tour = {}'.format(len(train_json), len(test_json)))
```

```python
def proccess_train(tourn_id, tourn):
    res = []
    for team in tourn:
        for member in team['teamMembers']:
            for quest_n, quest in enumerate(team['mask']):
                row = {}
                row['member_id'] = member['player']['id']
                row['team_id'] = team['team']['id']
                row['answer'] = quest
                row['answer_number'] = quest_n
                row['member_rating'] = member['rating']
                row['member_usedRating'] = member['usedRating']
                row['tournament_id'] = tourn_id
                row['position'] = team['position']
                row['questionsTotal'] = team['questionsTotal']
                res.append(row)
    return res
```

```python
train_df = pd.DataFrame(
    data=list(chain.from_iterable([proccess_train(key, value) for key, value in train_json.items()]))
)
train_df = train_df[(train_df['answer'] == '0') | (train_df['answer'] == '1')]
train_df['answer'] = train_df['answer'].astype(int)
train_df.head()
```

```python
def parse_tournaments(x):
    res = {}
    res['id'] = x['id']
    res['name'] = x['name']
    res['dateStart'] = dt.datetime.fromisoformat(x['dateStart']).date()
    res['type_id'] = x['type']['id']
    res['type_name'] = x['type']['name']
    res['season'] = x['season']
    res['questionQty'] = x['questionQty']
    return res


tournaments_json = pkl.load(open('./chgk/tournaments.pkl', 'rb'))

tournaments = pd.DataFrame(data=[parse_tournaments(x) for x in tournaments_json.values()])
tournaments = tournaments[tournaments['type_name'] != 'Общий зачёт']
tour_filt = set(train_df['tournament_id'])
print('Full: ', len(tournaments))
train_tournaments = tournaments[tournaments['id'].isin(tour_filt)].copy()
print('Train: ', len(train_tournaments))
```

```python
def proccess_questionQty(questionQty):
    keys = sorted(list(map(int, questionQty.keys())))
    res = []
    cum_sum = 0
    res.append(cum_sum)
    for key in keys:
        cum_sum += questionQty[str(key)]
        res.append(cum_sum)
    return res

train_tournaments['quest_n_cumsum'] = train_tournaments['questionQty'].apply(proccess_questionQty)
train_tournaments = pd.get_dummies(train_tournaments, columns=['type_name'], prefix='dumm_tn')
train_tournaments.head()
```

```python
def procces_quest_lap_n(data):
    ans_n, cum_sum = data
    lap_n = bisect.bisect_left(cum_sum, ans_n + 1)
    return lap_n


def procces_quest_first_quest_n(data):
    ans_n, cum_sum = data
    lap_n = bisect.bisect_left(cum_sum, ans_n + 1)
    return cum_sum[lap_n-1]


def procces_quest_last_quest_n(data):
    ans_n, cum_sum = data
    lap_n = bisect.bisect_left(cum_sum, ans_n + 1)
    return cum_sum[lap_n]


columns = list(train_df.columns)
columns = columns + [col for col in train_tournaments.columns if col.startswith('dumm_tn')]
train_df = train_df.merge(train_tournaments, left_on='tournament_id', right_on='id', how='inner')
quest_feat_col = ['lap_n', 'first_quest_n', 'last_quest_n']
# Честно не понимаю пандас, когда возрвращал сразу 3 колонки работало супер долго
train_df['lap_n'] = train_df[['answer_number', 'quest_n_cumsum']].progress_apply(procces_quest_lap_n, axis=1)
train_df['first_quest_n'] = train_df[['answer_number', 'quest_n_cumsum']].progress_apply(procces_quest_first_quest_n, axis=1)
train_df['last_quest_n'] = train_df[['answer_number', 'quest_n_cumsum']].progress_apply(procces_quest_last_quest_n, axis=1)
train_df['corr_ans'] = train_df['questionsTotal'] / train_df['quest_n_cumsum'].progress_apply(lambda x: x[-1])

columns = columns + quest_feat_col
train_df = train_df[columns]
train_df.head()
```

### Небольшое рассуждение по поводу EM

Судя по задаче с сурками, которая разбиралась на семинаре мы можем сделать тоже самое, только со следующими отличиями:
1) Поменяв 1 на 0, потому что в случае сурков мы не знали водяться они или нет при 0, а сейчас мы не знаем кто как отвечал в команде, но знаем что при неправильном ответе точно никто не знал ответ. По этому подходящей стратегией обучения будет будет инициализировать начальное прибледежение а потом итерироваться на каждом шагу насильно выставляя 0 там где команда не ответила

2) Избавившись от случайной велечины S потому что можно считать, что у нас нет перекоса

Плюс в конце хочется добваить что нам при составлении ранжированного списка, можно не учитывать константу перед сигмоидой и изменыннй свободный член внтури функции, потому что это просто гомотетия и сдвиг

```python
import torch

class LogisticRegression(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegression, self).__init__()
        self.linear = torch.nn.Linear(input_dim, output_dim)
    def forward(self, x):
        outputs = torch.sigmoid(self.linear(x))
        return outputs
```

```python
model = LogisticRegression(len(x_cols),1)
criterion = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)
```
